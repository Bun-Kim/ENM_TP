# TP4 : Extrêmes

## Introduction

---

Dans ce TP nous allons travailler sur les extrêmes, qui suivent la loi dites $\mathrm{GEV}(\mu,\sigma,\xi)$ (*Generalized Extreme Value*). Elle admet trois paramètres:

- $\mu$, paramètre de position, similaire à la moyenne (mais pas exactement),
- $\sigma$, paramètre d'échelle, similaire à l'écart-type (mais pas exactement),
- $\xi$, paramètre de forme.

Elle admet comme fonction de répartition (en anglais *Cumulative Distribution Function*, cdf):

$$G : x\mapsto \exp\left[-\left(1+\xi\frac{x-\mu}{\sigma}\right)^{-1/\xi}\right]$$

Et comme fonction quantile (l'inverse de la fonction de répartition):

$$\mathcal{Q}:p\mapsto \mu +\frac{\sigma}{\xi}\left[ \left(-\log p\right)^{-\xi} -1 \right]$$

Une fonction pour controler les tailles des figures:

```r
## Cette fonction permet d'initialiser une nouvelle figure, composé de nrow * ncol sous figures
## nrow: nombre de lignes de sous-figures
## ncol: nombre de colones de sous-figures
## ratio: ratio du nombre de pixel entre la largeur et la hauteur de la figure
## scale: multiplicateur du nombre de pixel en hauteur et largeur
new_fig = function( nrow = 1 , ncol = 1 , ratio = 4 / 3 , scale = 7 )
{
	base::options( repr.plot.width = ratio * scale * ncol , repr.plot.height = scale * nrow )
	graphics::par( mfrow = base::c(nrow,ncol) ) ## On construit un graphique découpé en nrow * ncol fenêtres
}
```

Deux fonctions à usage interne (d'après vous à quoi servent-elles ?):

*Indication : dans les deux cas $p=G(x)$*

```r
qgev_gradient = function( p , gev , lower.tail = TRUE )
{
    if(!lower.tail)
        p = 1 - p
    loc   = gev$loc
    scale = gev$scale
    shape = gev$shape
    tp    = -log(p)
    gradloc   = 1
    gradscale = ( tp^(-shape) - 1 ) / shape
    gradshape = scale * ( (1 - tp^(-shape)) / shape^2 - tp^(-shape) * log(tp) / shape )
    grad = cbind(gradloc,gradscale,gradshape)
    
    return(grad)
}
```

```r
pgev_gradient = function( x , gev , lower.tail = TRUE )
{
    loc   = gev$loc
    scale = gev$scale
    shape = gev$shape
    
    p    = gev$cdf(x)
    Z    = (x-loc) / scale
    ZZ   = 1 + shape * Z
    ZZs  = ZZ^(-1/shape)
    ZZs1 = ZZs / ZZ
    
    gradloc   = -p * ZZs1 / scale
    gradscale = -p * ZZs1 * Z / scale
    gradshape = -p / shape^2 * ZZs * ( log(ZZ) - shape * Z / ZZ )
    grad = cbind(gradloc,gradscale,gradshape)
    if(!lower.tail)
        grad = -grad
    
    return(grad)
}
```

Une fonction pour les niveaux de retours et leurs intervalles de confiances:

```r
## Calcul des niveaux de retour et de leurs intervalles de confiances
## par la méthode delta
## x: quantiles
## gev: loi GEV fitté
## Rt: si p est une durée de retour ou une proba
qgev_CI = function( p , gev , Rt = FALSE , alpha = 0.05 )
{
    if(Rt) p = 1. / p
    grad = as.matrix( qgev_gradient( p , gev , FALSE ) , ncol = 3 , nrow = length(p) )
    qvar = colSums(base::t(grad) * (gev$cov %*% base::t(grad)))
    names(qvar) = c()
    
    qlevel       = gev$isf(p)
    qlevel_left  = qlevel + stats::qnorm(     alpha / 2 ) * base::sqrt(qvar)
    qlevel_right = qlevel + stats::qnorm( 1 - alpha / 2 ) * base::sqrt(qvar)
    
    return( list(value = qlevel , left = qlevel_left , right = qlevel_right ) )
}
```

Une fonction pour les durées de retours et leurs intervalles de confiances:

```r
## Calcul des durées de retour et de leurs intervalles de confiances
## par la méthode delta
## x: quantiles
## gev: loi GEV fitté
## Rt: renvoi (ou non) le résultat sous forme de durée de retour
pgev_CI = function( x , gev , Rt = FALSE , alpha = 0.05 )
{
    grad = as.matrix( pgev_gradient( x , gev , FALSE ) , ncol = 3 , nrow = length(x) )
    pvar = colSums(base::t(grad) * (gev$cov %*% base::t(grad)))
    names(pvar) = c()
    
    plevel       = gev$sf(x)
    plevel_left  = plevel + stats::qnorm(     alpha / 2 ) * base::sqrt(pvar)
    plevel_right = plevel + stats::qnorm( 1 - alpha / 2 ) * base::sqrt(pvar)
    plevel_left  = pmax( 0 , plevel_left )
    plevel_left  = pmin( 1 , plevel_left )
    plevel_right = pmax( 0 , plevel_right )
    plevel_right = pmin( 1 , plevel_right )
    
    if(Rt)
    {
        plevel = 1. / plevel
        left   = 1. / plevel_right
        right  = 1. / plevel_left
        plevel_left  = left
        plevel_right = right
    }
    
    return( list(value = plevel , left = plevel_left , right = plevel_right ) )
}
```

Et une dernière pour un diagnostique global:

```r
## Fonction de diagnostique d'une loi GEV
## Y: jeu de donnée
## GEV: Un objet ROOPSD::GEV fitté
## Exemple:
## >>> gev = GEV$new()$fit(Y)
## >>> gev_diag( Y , gev )
gev_diag = function( Y , gev )
{
    rvY   = rv_histogram$new(Y = Y)
    loc   = gev$loc
    scale = gev$scale
    shape = gev$shape

    new_fig(2,2,scale=5)

    ## Probability plot
    emp = rvY$cdf(Y)
    mod = gev$cdf(Y)
    plot( emp , mod , xlab = "Empirical" , ylab = "Model" , xlim = base::c(0,1) , ylim = base::c(0,1) , main = "Probability" )
    lines( c(0,1) , c(0,1) , col = "blue" )
    
    ## Quantile plot
    p   = seq( 0.01 , 0.99 , length = 100 )
    emp = rvY$icdf(p)
    mod = gev$icdf(p)
    xylim = base::c(min(emp,mod),max(emp,mod))
    plot( mod , emp , xlim = xylim , ylim = xylim , xlab = "Model" , ylab = "Empirical" , main = "Quantile ")
    lines( xylim , xylim , col = "blue" )
    
    ## Return level plot
    p   = gev$cdf(Y)
    Rts = 10^base::seq( base::log10( base::min(1/p) )  , base::log10( base::max(1/p) ) , length = 100 )
    p   = 1. / Rts
    qlevel = qgev_CI( p , gev , FALSE )
    plot(  Rts , qlevel$value , type = "l" , log = "x" , main = "Return level" , xlab = "Return time" , ylab = "Level" )
    lines( Rts , qlevel$left  , col = "blue" )
    lines( Rts , qlevel$right , col = "blue" )
    points( 1. / rvY$sf(Y) , Y )

    ## Density plot
    minY = base::min(Y)
    maxY = base::max(Y)
    delY = 0.1 * (maxY-minY)
    x = base::seq( minY - delY , maxY + delY , length = 1000 )
    hist( Y , breaks = floor(0.1*length(Y)) + 1 , col = grDevices::rgb(1,0,0,0.1) , freq = FALSE , xlab = "Value" , main = "Density" )
    lines( x , gev$density(x) , col = "red" , type = "l" )
    points( Y , base::rep(0,length(Y)) )
    
    invisible(NULL)
}
```

### A.1 Fit non stationaire d'une loi GEV

```r
data = read.csv( "ToulouseBlagnac_TNn_TXx.csv" , header = TRUE , sep = ";" )
```

Un problème posé par le changement climatique est qu'il entraine un changement des paramètres $\mu$ et $\sigma$ au cours du temps (peut-être même $\xi$!). Nous allons voir ici une manière de régler le problème de $\mu$, qui repose sur le résultat suivant: si $T\sim\mathrm{GEV}(\mu,\sigma,\xi)$, alors:

$$T-\mu\sim\mathrm{GEV}(0,\sigma,\xi)$$

Partant de la, nous allons commencer par estimer la tendance dans le jeu de données `TXx`. Pour cela nous allons tout d'abord lisser la variable `TXx`, puis effectuer une regression linéaire entre la variable lissée et `TXx`, ce qui nous donnera une nouvelle variable nommée `trend`:

```r
spl    = smooth.spline( data$time , data$TXx , df = 3 )
res_lm = lm( data$TXx ~ spl$y )
trend  = predict(res_lm)

new_fig(1,1)
plot( data$time , data$TXx , xlab = "Time" , ylab = "Temperature" )
lines( data$time , trend , col = "red" )
```

L'hypothèse que nous ferons est que le paramètre $\mu$ s'écrit:
$$\mu = \mu_0 + \mathrm{trend}$$
Et on obtient:
$$T-\mathrm{trend}\sim\mathrm{GEV}(\mu_0,\sigma,\xi)$$

Autrement dit, nous avons retiré l'effet du changement climatique des données. Nous pouvons maintenant estimer cette nouvelle loi:

```r
TXx_s  = data$TXx - trend
gev = GEV$new()$fit(Y = TXx_s)
gev_diag( TXx_s , gev )
```

Et notre jeu de donnée d'origine suivra une loi GEV avec les paramètres suivants (attention à $\mu$ qui change au cours du temps):

```r
n_time = length(data$time)
loc   = gev$loc + trend
scale = base::rep( gev$scale , n_time )
shape = base::rep( gev$shape , n_time )
```

Nous pouvons estimer la durée de retour au cours du temps de la valeur 42°C, ou le quantile associé à la durée de retour 30 ans:

```r
p = pgev( base::rep( 42 , n_time ) , loc = loc , scale = scale , shape = shape , lower.tail = FALSE )
q = qgev( base::rep( 1 / 30 , n_time ) , loc = loc , scale = scale , shape = shape , lower.tail = FALSE )
```

```r
new_fig(1,2)
plot( data$time , p , log = "y" , type = "l" , xlab = "Time" , ylab = "Probabilité" )
plot( data$time , q , type = "l" , xlab = "Time" , ylab = "Quantile")
```

Nous pouvons aussi comparer les probas / quantiles à leurs moyennes sur la période 1947 / 1977:

```r
new_fig(1,2)
plot( data$time , p / mean(p[1:30]) , log = "y" , type = "l" , xlab = "Time" , ylab = "Changement de risque" )
plot( data$time , q - mean(q[1:30]) , type = "l" , xlab = "Time" , ylab = "Changement de temp.")
```

Nous pouvons voir que la probabilité d'avoir au moins 42°C a été multiplié par $\sim 15$, tandis que la température associé à la durée de retour 30 ans a augmenté de $\sim 2.5\mathrm{K}$ par rapport à la moyenne 1947 / 1977.

- La probabilité d'avoir au moins 42°C a augmenté de 30%,
- L'intensité d'un tel évènement a augmenté de 0.27K.

### A.2 Gradient utilisé par la méthode delta pour le calcul des intervalles de confiances

#### A.2.1 Gradient de la fonction de répartition d'une GEV

Avec les notations suivantes:

$$G(x) = \exp\left(-\left[1+\xi\left(\frac{x-\mu}{\sigma}\right)\right]^{-1/\xi}\right) = \exp\left(-\mathcal{Z}^{-1/\xi}\right)$$

$$
\begin{aligned}
Z &= \frac{x-\mu}{\sigma}\\
\mathcal{Z} &= 1+\xi Z
\end{aligned}
$$


On a pour $\partial_\mu$:
$$
\begin{aligned}
\partial_\mu G(x) &= \partial_\mu\exp\left(-\mathcal{Z}^{-1/\xi}\right)\\
&= -G(x)\partial_\mu\left(\mathcal{Z}^{-1/\xi}\right) = -G(x)\partial_\mu\left(e^{-\log(\mathcal{Z})/\xi}\right)\\
&= -G(x)\mathcal{Z}^{-1/\xi}\left(-\frac{1}{\xi}\right)\frac{\partial_\mu \mathcal{Z}}{\mathcal{Z}}\\
&= -G(x)\mathcal{Z}^{-1/\xi-1}\left(-\frac{1}{\xi}\right)\left(-\frac{\xi}{\sigma}\right)\\
&= -G(x)\mathcal{Z}^{-1/\xi-1}\frac{1}{\sigma}
\end{aligned}
$$

On a pour $\partial_\sigma$:
$$
\begin{aligned}
\partial_\sigma G(x) &= \partial_\sigma\exp\left(-\mathcal{Z}^{-1/\xi}\right)\\
&= -G(x)\partial_\sigma\left(\mathcal{Z}^{-1/\xi}\right) = -G(x)\partial_\sigma\left(e^{-\log(\mathcal{Z})/\xi}\right)\\
&= -G(x)\mathcal{Z}^{-1/\xi}\left(-\frac{1}{\xi}\right)\frac{\partial_\sigma \mathcal{Z}}{\mathcal{Z}}\\
&= -G(x)\mathcal{Z}^{-1/\xi-1}\left(-\frac{1}{\xi}\right)\left(-\frac{\xi(x-\mu)}{\sigma^2}\right)\\
&= -G(x)\mathcal{Z}^{-1/\xi-1}\frac{Z}{\sigma}
\end{aligned}
$$

On a pour $\partial_\xi$:
$$
\begin{aligned}
\partial_\xi G(x) &= \partial_\xi\exp\left(-\mathcal{Z}^{-1/\xi}\right)\\
&= -G(x)\partial_\xi\left(\mathcal{Z}^{-1/\xi}\right) = -G(x)\partial_\xi\left(e^{-\log(\mathcal{Z})/\xi}\right)\\
&= -G(x)\mathcal{Z}^{-1/\xi}\partial_\xi\left(-\frac{\log\mathcal{Z}}{\xi}\right)\\
&= -G(x)\mathcal{Z}^{-1/\xi}\left(-\frac{\xi\frac{\partial_\xi\mathcal{Z}}{\mathcal{Z}}-\log\mathcal{Z}}{\xi^2}\right)\\
&= -\frac{1}{\xi^2}G(x)\mathcal{Z}^{-1/\xi}\left(\log\mathcal{Z}-\xi\frac{Z}{\mathcal{Z}}\right)
\end{aligned}
$$

#### A.2.2 Gradient de la fonction quantile d'une GEV

$$\mathcal{Q}(p)=\mu+\frac{\sigma}{\xi}\left[\left(-\log p\right)^{-\xi}-1\right]$$

$$
\begin{aligned}
\partial_\mu\mathcal{Q}(p) = 1
\end{aligned}
$$

$$
\begin{aligned}
\partial_\sigma\mathcal{Q}(p) = \frac{1}{\xi}\left[\left(-\log p\right)^{-\xi}-1\right]
\end{aligned}
$$

$$
\begin{aligned}
\partial_\xi\mathcal{Q}(p) &= \sigma\partial_\xi\left[\frac{1}{\xi}\left[\left(-\log p\right)^{-\xi}-1\right]\right]\\
&= \sigma\partial_\xi\left[\frac{1}{\xi}\left[\tilde{p}^{-\xi}-1\right]\right]\\
&= \sigma\left[-\frac{1}{\xi^2}\left[\tilde{p}^{-\xi}-1\right]+\frac{1}{\xi}\partial_\xi\tilde{p}^{-\xi}\right]\\
&= \sigma\left[-\frac{1}{\xi^2}\left[\tilde{p}^{-\xi}-1\right]-\frac{1}{\xi}\log(\tilde{p})\tilde{p}^{-\xi} \right]
\end{aligned}
$$
